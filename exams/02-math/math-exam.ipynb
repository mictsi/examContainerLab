```python
from ipywidgets import interact

@interact(k=(1, 10, 1))
def bode(k=3):
    xs = np.linspace(0, 2 * np.pi, 400)
    plt.plot(xs, np.sin(k * xs))
    plt.title(f'sin({k}x)')
    plt.grid(True)
    plt.show()
```
